```python
# Q4

numbers = [88,5,7,99,55,65]
l = len(numbers) - 1
s = 0

while s < l:
    if numbers[s] > numbers[l]:
        numbers[s], numbers[l] = numbers[l], numbers[s]

    l -= 1

    if s == l:
        l = len(numbers) - 1
        s += 1

print("Ascending list:", numbers)
```

```python
# Q7
nested_list = [1, [2, 3], [4, [5, 6], 7], 8]

fla_list = []

def flat(nested_list):
  for i in nested_list:
    if (type(i)==list):
      flat(i)
    else:
      fla_list.append(i)
flat(nested_list)
print(fla_list)


```

```python
## Intermediate
#Q9
num = [1, 2, 3, 2, 1]
i=0
li=len(num)-1
pal=[]
while i<len(num):
  if (num[i] == num [li]):
    pal.append(num[i])
  li-=1
  i+=1
if (pal == num):
  print("Palindrome: ",pal)
else:
  print("Not a Palindrome")
```

```python
# Q5
num = [-1, 0, 1, 2, -1, -4]
n = len(num)
result = []

for i in range(n):
  for j in range(0,n-i-1):
    if num[j]>num[j+1]:
      temp = num[j]
      num[j]=num[j+1]
      num[j+1]=temp
for i in range(n):
    if i > 0 and num[i] == num[i-1]:
        continue
    left = i + 1
    right = n - 1
while left < right:
        s = num[i] + num[left] + num[right]

        if s == 0:
            result.append((num[i], num[left], num[right]))
            left += 1
            right -= 1

        elif s < 0:
            left += 1
        else:
            right -= 1

print(result)

print(num)


```
